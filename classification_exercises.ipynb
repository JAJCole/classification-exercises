Create a function named prep_iris that accepts the untransformed iris data, and returns the data with the transformations above applied.

```python
def prep_iris():
    iris = get_iris_data()
    iris = iris.drop(columns=['species_id','measurement_id','Unnamed: 0'])
    iris = iris.rename(columns= {'species_name':'species'})
    dummy_iris = pd.get_dummies(iris.species, drop_first=True)
    iris = pd.concat([iris, dummy_iris], axis=1)
    return iris
```

# Split your data

Write a function to split your data into train, test and validate datasets. Add this function to prepare.py. 

```python
# a good data split is 60/20/20 with 60 being train

# ranges = df_excel[df_excel_nums].max() - df_excel[df_excel_nums].min()
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python


def my_train_test_split(df, target=None):
    if target:
        train, test = train_test_split(df, test_size=.2, random_state=123, stratify=df[target])
        train, validate = train_test_split(train, test_size=.25, random_state=123, stratify=train[target])
    else:
        train, test = train_test_split(df, test_size=.2, random_state=123, stratify=df[target])
        train, validate = train_test_split(train, test_size=.25, random_state=123, stratify=train[target])
    return train, validate, test

```

# Bivariate Stats

Visualize each measurement type (y-axis) with the species variable (x-axis) using barplots, adding a horizontal line showing the overall mean of the metric (y-axis). 

```python
import seaborn as sns
```

```python

sns.barplot(x='species', y='sepal_length', data=train_iris)
avg_sep_len = train_iris.sepal_length.mean()
plt.axhline(avg_sep_len, label='avg_sepal_length', color='r')
plt.legend()
plt.show()
```

```python
# function for all features and their avg line


features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

plot_avg_line = lambda feature: (
    sns.barplot(x='species', y=feature, data=train_iris),
    plt.axhline(train_iris[feature].mean(), label=f'avg_{feature}', color='r'),
    plt.legend(),
    plt.show()
)

for feature in features:
    plot_avg_line(feature)
```

For virginica & versicolor: Compare the mean petal_width using the Mann-Whitney test (scipy.stats.mannwhitneyu) to see if there is a significant difference between the two groups. Do the same for the other measurement types. 

```python
from scipy import stats
```

```python
virginica = train_iris[train_iris.species == 'virginica']
versicolor = train_iris[train_iris.species == 'versicolor']
setosa = train_iris[train_iris.species == 'setosa']

features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    t, p = t, p = stats.mannwhitneyu(virginica[feature], versicolor[feature], alternative="two-sided")
    print('Comparing {} between virginica and versicolor'.format(feature))
    print('t:', t)
    print('p:', p)
    print('\n')

```
